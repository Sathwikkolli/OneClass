```python
if __name__ == "__main__":
    import torch
    import pickle
    import os

    # Device setup
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Create the detector (do NOT retrain)
    model_name = "wavlm_large"
    speaker = "Donald_Trump"
    detector = AudioSVMClassifier(speaker, model_name, device)

    # Load previously saved scaler and svm_model
    saved_model_dir = "/data/anomaly_detection/single_class/W_svm_trained_models/Donald_Trump/"
    scaling_model = os.path.join(saved_model_dir, 'scaling_models', f"{model_name}" + '.pkl')
    svm_model = os.path.join(saved_model_dir, 'svm_models', f"{model_name}" + '.pkl')

    # Load the scaler and SVM model
    detector.scaler = joblib.load(scaling_model)
    detector.svm_model = joblib.load(svm_model)

    # Use the model to predict
    test_audio_file = "/home/selloh/speaker-specific-ASD/donald_trump.mp3"
    prediction = detector.predict_from_audio(test_audio_file, load_model_dir=None)  # don't reload here again
    print("Prediction for the test audio file:", prediction)
```
